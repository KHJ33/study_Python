## 내용 추출

```python
def crolling(df) :
    univ = driver.find_element(By.ID, 'm_title3').text[1:-1]

    tbody = driver.find_element(By.ID, 'listView3')
    tr = tbody.find_elements(By.TAG_NAME, 'tr')
    
    val = []
    for i in tr :
        td = i.find_elements(By.TAG_NAME, 'td')
        temp = []
        temp.append(univ)
        for j in td :
            temp.append(j.text)

        val.append(temp)
    df1 = pd.DataFrame(val)
    
    df = pd.concat([df, df1])
    
    return df
```

```python
df = pd.DataFrame()
# df = crolling(df)
for _ in range(133) :
    if _ == 0 :
        select_year = Select(driver.find_element(By.NAME, 'EvalYear'))
        select_year.select_by_index(1)
    
    select_univ = Select(driver.find_element(By.NAME, 'EvalUniv'))
    select_univ.select_by_index(_)
    time.sleep(0.5)
    # 전체 색택
    if _ == 0 :
        driver.find_element(By.CLASS_NAME, 'multiselect-dropdown-arrow').click()
    #     time.sleep(0.5)
        driver.find_element(By.CLASS_NAME, 'multiselect-checkbox').click()
    #     time.sleep(0.5)
    
    select_search = driver.find_element(By.CLASS_NAME, 'btn_search')
    select_search.click()
    
    time.sleep(0.5)
    df = crolling(df)
```
