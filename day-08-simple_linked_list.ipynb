```python
# Node 생성과 node 출력을 함수로 정의해서 구현
class Node :
    def __init__(self) :
        self.data = None
        self.next = None
        
def add_node(node) : # 마지막에 노드 추가
    curr_node = head
    
    while curr_node.next :
        curr_node = curr_node.next
    curr_node.next = node

def print_node() : # 전체 자료 출력
    curr_node = head
    while curr_node :
        print(curr_node.data, end = ' \t')
        curr_node = curr_node.next

head = None # node 가 존재하지 않음 처음 위치 설정
while True :
    data = input('추가할 노드 데이터 입력 : ')
    if data == 'q' :
        break
    
    node = Node() # Node 생성
    node.data = data
    
    if head == None :
        head = node
    else :
        add_node(node) # 생성된 노드를 기존의 자료의 마지막에 추가
        
print_node()
```

```python
# 삽입시 정렬 되어 node 구성
class Node :
    def __init__(self) :
        self.data = None
        self.next = None
        
def sort_add_node(new_node) : # 마지막에 노드 추가
    global head
    now_node = head
    
    if new_node.data < now_node.data :
        new_node.next = now_node
        head = new_node
    
    else :
        while now_node.next :
            if new_node.data < now_node.next.data :
                new_node.next = now_node.next
                now_node.next = new_node
                return
            now_node = now_node.next
            
        now_node.next = new_node
    
def print_node() : # 전체 자료 출력
    curr_node = head
    while curr_node :
        print(curr_node.data, end = ' \t')
        curr_node = curr_node.next
        
def delete_node(data) :
    global head
    now_node = head
    
    if data == now_node.data :
        head = now_node.next
        del now_node
    else :
        while now_node.next :
            if now_node.next.next == None and now_node.next.data == data:
                temp = now_node.next
                now_node.next = None
                del temp
                break
            if data == now_node.next.data :
                temp = now_node.next
                now_node.next = temp.next
                del temp
                break
            now_node = now_node.next
```

```python
head = None

while True :
    data = input('추가할 노드 데이터 입력 : ')
    
    if data == 'q' :
        break
    
    new_node = Node()
    new_node.data = data
    
    if not head :
        head = new_node
    else :
        sort_add_node(new_node)

print_node()
```

```python
# 이름 성적을 입력 받아 선형연결로 데이터 저장 -> 순서 없이 마지막에 연결

# 1. 검색 하고자 하는 이름 입력을 받아 해당 자료를 검색 후 이름과 점수 출력
# 2. 입력된 자료를 딕셔너리(이름을 키로 성적을 데이터) 로 저장 후 
# 3. 성적이 가장 높은 학생의 이름과 성적 출력
# 4. 3번에서 찾은 학생의 자료를 삭제 -> 딕셔너리와 링크 자료 둘 다 삭제

class Student :
    def __init__(self, name , score) :
        self.name = name
        self.score = score
        self.next = None
        
def create_link_data (new_node) :
    global head
    now_node = head
    
    if not head :
        head = new_node
    else :
        while now_node.next :
            now_node = now_node.next
        now_node.next = new_node
    
def search_name(name) :
    global head
    now_node = head

    while now_node :
        if now_node.name == name :
            return f'검색된 {name} 의 성적은 {now_node.score} 입니다.'
        now_node = now_node.next
    return f'검색된 {name} 은 존재하지 않습니다.'

def link_data_to_dict() :
    global head
    dic_temp = {}
    now_node = head
    
    while now_node :
        dic_temp[now_node.name] = now_node.score
        now_node = now_node.next
    return dic_temp

def search_student_max() :
    global head
    now_node = head
    max_score = 0
    
    while now_node :
        if max_score < now_node.score :
            max_score = now_node.score
        now_node = now_node.next
        
    return max_score

def delete_data (max_score):
    global head
    global dic_student
    now_node = head
    
    for key, value in dic_student.copy().items() :
        if value == max_score :
            del dic_student[key]
            
    while now_node :
        if now_node.score == max_score :
            if now_node == head :
                temp = now_node
                now_node = now_node.next
                head = head.next
                del temp
            else :
                before_node.next = now_node.next
                temp = now_node
                now_node  = now_node.next
                del temp
        else :        
            before_node = now_node
            now_node = now_node.next

def print_data (max_score = 0) :
    curr_node = head
    if max_score == 0 :
        while curr_node :
            print(curr_node.name, curr_node.score, end = ' \t')
            curr_node = curr_node.next
    else :
        while curr_node :
            if max_score == curr_node.score :
                print(curr_node.name, curr_node.score, end = ' \t')
            curr_node = curr_node.next
    print()
```

```python
head = None
dic_student = {}

while True :
    input_info = input('이름과 성적을 입력하세요 : ').split()
    
    if input_info[0] == 'q' :
        break
    if len(input_info) == 2 and input_info[0].isalpha() and input_info[1].isdigit() :
        new_node = Student(input_info[0], int(input_info[1]))
        create_link_data(new_node)
    else :
        print('입력을 확인하세요')
        
dic_student = link_data_to_dict()

print('\n\n')
name = input('검색하고자 하는 이름을 입력하시오 : ')
result = search_name(name)
print(result)

print('\n\n')
print('성적이 가장 높은 학생의 정보 : ', end = '\t')
max_score = search_student_max()
print_data(max_score)

print('\n\n')
print('성적이 가장 높은 학생의 정보를 삭제 전', end ='\t')
print_data()
delete_data(max_score)
print('성적이 가장 높은 학생의 정보를 삭제 후', end = '\t')
print_data()
```

```python
class Node() :
    def __init__(self) :
        self.data = None
        self.perv = None
        self.next = None
        
def insert_node(new_node) :
    global HEAD
    global LAST
    
    now_node = HEAD
    
    if new_node.data < now_node.data :
        new_node.next = now_node
        now_node.perv = new_node
        HEAD = new_node
        
    else :
        while now_node.next :
            if new_node.data < now_node.next.data :
                new_node.next = now_node.next
                new_node.perv = now_node
                
                now_node.next.perv = new_node
                now_node.next = new_node
                return
                
            now_node = now_node.next
        
        now_node.next = new_node
        new_node.perv = now_node
        LAST = new_node

def delete_node(name) :
    global HEAD
    global LAST
    now_node = HEAD
    
    if now_node.data == name :
        HEAD = now_node.next
        now_node.next.perv = None
        del now_node
        
    else :
        while now_node.next :
            if now_node.data == name :
                now_node.next.perv = now_node.perv
                now_node.perv.next = now_node.next
                del now_node
                return
            now_node = now_node.next
        
        if now_node.data == name :
            now_node.perv.next = None
            LAST = now_node.perv
            del now_node

def print_node() :
    global HEAD
    global LAST
    
    now_node = HEAD
    
    print('순방향으로 출력 : ', end = '\t')
    
    while now_node :
        print(f'{now_node.data}', end = '\t')
        now_node = now_node.next
        
    print()
    print('역방향으로 출력 : ', end = '\t')
    
    now_node = LAST
    while now_node :
        print(f'{now_node.data}', end = '\t')
        now_node = now_node.perv
        
```

```python
HEAD = None
LAST = None

while True :
    name = input('이름을 입력하세요 : ')
    
    if name == 'q' :
        break
    
    new_node = Node()
    new_node.data = name
    
    if not HEAD :
        HEAD = new_node
        LAST = new_node
    else:    
        insert_node(new_node)

print_node()
print()
# name = input('삭제할 이름을 입력하세요 : ')
delete_node(name)
print_node()
```
