```python
import random

random.randint(1, 45)

# 1 부터 45까지의 수중에서 로또번호 6개를 출력하는 프로그램 작성
# lotto_func() 을 정의해서 작성
# 입력한 횟수만큼 반복 : 입력 횟수를 키보드로 부터 받음

def lotto_func() :
    total_lotto = []
    
    while True :
        lotto_num = random.randint(1, 45)
        
        if lotto_num in total_lotto :
            pass
        
        else :
            total_lotto.append(lotto_num)
            
        if len(total_lotto) == 6 :
            total_lotto.sort()
            return total_lotto

total_lotto = []
num = int(input('반복 횟수를 입력하시오 : '))

for i in range(num) :
    total_lotto = lotto_func()
    print(f'{i + 1} 번째 로또 번호는', *total_lotto, '입니다.')
```

```python
poet = '''
나 보기가 역겨워 가실 때에는 말없이 고이 보내 드리오리다
영변에 약산 진달래꽃 아름 따다 가실 길에 뿌리오리다
가시는 걸음 걸음 놓인 그 꽃을 사뿐히 즈려 밟고 가시옵소서
나 보기가 역겨워 가실 때에는 죽어도 아니 눈물 흘리오리다
'''

# 각 글자의 빈도수를 출력 : 빈도수가 3 이상인 글자만 출력

char_dict = {}

for char in poet:
    if char.isalpha() :
        if char in char_dict : # char 가 char_dict 에 존재하는지 확인
            char_dict[char] += 1
        else :
            char_dict[char] = 1
        
for key, value in char_dict.items() : # 키와 값을 가져 옴
    if value >= 3 : # 빈도수가 3 이상이면 출력
        print(f'{key} : {value}', end = '\t')
```

# ✪✪✪✪✪

```python
def power(x) :
    return x * x

def under_3(x) :
    return x < 3

x_list = [1, 2, 3, 4, 5]
# map(함수, 리스트) 함수를 이용하여 x_list의 각 요소에 power() 함수 실행

output_a = map(power, x_list) # power() 함수 호출 -> 리스트의 각 요소별로 호출
print('map() 함수 실행 결과', output_a)
# output_a = [ x * x for x in x_list ]
print('map() 함수 실행 결과', list(output_a))

# filter(함수, 리스트) 함수를 이용하여 x_list의 각 요소에 under_3() 함수 실행

output_b = filter(under_3, x_list)
# output_b = [ x for x in x_list if x < 3 ]
print('filter() 함수 실행 결과', list(output_b))
```

```python
# lambda 매개변수 : return 
output_a = map(lambda x : x * x, x_list)
output_b = filter(lambda x : x < 3 , x_list)

print('map() 함수 실행 결과', list(output_a))
print('filter() 함수 실행 결과', list(output_b))
```

```python
# 저장된 자료를 읽어 와서 이름, 몸무게, 키, BMI 출력을 통한 정상 체중(과체중, 저체중)
# BMI = (몸무게 / (키  ** 2) )

with open('./info.txt', 'r') as file :
    for line in file :
        line = line.strip()
        name , weight , height = line.split(', ')
        bmi = int(weight) / ((int(height) / 100) ** 2)
        if bmi < 20 :
            result = '저체중'
        elif bmi <= 24 :
            result = '정상체중'
        elif bmi <= 29 :
            result = '과체중'
        else :
            result = '비만'
        
        print(f'이름 : {name} 체중 : {weight} 키 : {height} BMI : {bmi:.2f} {result}')
```

```python
books = [ {'제목' : '파이썬' , '가격' : 18000} ,
         {'제목' : '머신러닝' , '가격' : 20000} , 
         {'제목' : '자바스크립트' , '가격' : 24000}
        ]

# 가격 추출 함수
def price_output(book) :
    return book['가격']

# 가장 가격이 저렴한 책
print(min(books, key = price_output))

# 가장 가격이 비싼 책
print(max(books, key = price_output))

# 가장 가격이 저렴한 책
print(min(books, key = lambda book : book['가격']))

# 가장 가격이 비싼 책
print(max(books, key = lambda book : book['가격']))
```

## chapter 6-1 예외처리
- try ~ except ~ else ~ finally

```python
try :
    num = int(input('숫자 입력 :'))
except :
    print('정수 입력 :')
else :
    print('원의 반지름 : ', num)
    print('원의 둘래 : ', num * 3.14 * 2)
    print('원의 넓이 : ', num * num * 3.14)
finally :
    print('end')
```

```python
# finally 구문 사용 예

try: 
    file = open('info.txt', 'r')
    file.read()
    file.write('test')

except Exception as e :
    print(e)

finally :
    print('finally --')
    file.close()
file.closed
```

```python
def test() :
    print('test func start ---')
    
    try :
        print('try 구문 실행')
        return 
        print('try return 실행')
        
    except :
        print('except 구문 실행')
    
    finally :
        print('finally 실행')
        
    print('test func end ---')
    
test()
```

```python
# 반목문에서 사용 -> break 실행시 finally 무조건 실행

while True :
    try :
        print('try 구문 실행')
        break
        print('try break 실행')
        
    except :
        print('except 구문 실행')
    
    finally:
        print('finally 실행') # 무조건 실행
    
    print('while end ---')

print('program end')
```

```python
# 변수를 선언합니다.
list_number = [52, 273, 32, 72, 100]

# try except 구문으로 예외를 처리합니다.
try:
    # 숫자를 입력 받습니다.
    number_input = int(input("정수 입력> "))
    # 리스트의 요소를 출력합니다.
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
#     예외.발생해주세요()
except ValueError as exception:
    # ValueError가 발생하는 경우
    print("정수를 입력해 주세요!")
    print(type(exception), exception)
except IndexError as exception:
    # IndexError가 발생하는 경우
    print("리스트의 인덱스를 벗어났어요!")
    print(type(exception), exception)
except Exception as exception:
    # 이외의 예외가 발생한 경우
    print("미리 파악하지 못한 예외가 발생했습니다.")
    print(type(exception), exception)
```
