```python
# queue 의 사이즈 입력
# e 또는 E : 데이터 삽입 enqueue()
# d 또는 D : 데이터 추출 dequeue()
# p 또는 P : queue 출력
# front , rear 변수 사용
```

```python
def enqueue(data) :
    global rear , queue
    
    rear += 1
    queue[rear] = data

def dequeue() :
    global front, rear , queue
    
#     if front + 1 > rear :
#         print('추출 가능한 데이터가 없습니다.')
#         return False
    
    front += 1
    data = queue[front]
    queue[front] = None
    
    return data

def print_queue() :
    global front , rear , queue_size
    
    for i in range(front + 1 , rear + 1 , 1) :
        print(queue[i] , end = ' , ')
    print()

def isQueueFull() :
    global rear , queue_size
    
    if rear == queue_size - 1 :
        return True
    else :
        return False

def isQueueEmpty() :
    global front , rear , queue
    
    if front == rear and queue[rear] == None :
        return True
    else :
        False

queue_size = int(input('Q 의 사이즈를 입력하세요 : '))
queue = [None for _ in range(queue_size)]
front = -1
rear = -1

if __name__ == '__main__' :
    
    while True :
        menu = input('e,E : 데이터 삽입 // d,D : 데이터 추출 // p,P : queue 출력 // q : 종료 // : \n')
        
        if menu == 'q' :
            break
        
        if menu == 'e' or menu == 'E' :
            if isQueueFull() :
                print('Q 는 꽉 차있습니다.')
            else :
                data = input('삽입 데이터를 입력하세요 : ')
                enqueue(data)
        elif menu == 'd' or menu == 'D':
            if isQueueEmpty() :
                print('추출 가능한 데이터가 없습니다.')
            else :
                data = dequeue()
                if data :
                    print('추출한 데이터 ', data)
        elif menu == 'p' or menu == 'P' :
            print_queue()
        else :
            print('menu 선택 오류')
```

```python
front = -1
rear = -1
queue_size = 10
queue = [None for _ in range(queue_size)]

def waiting(guest) :
    global front , rear , queue_size , queue
    
    if rear + guest > queue_size -1:
        return False
    
    for _ in range(guest) :
        rear += 1
        queue[rear] = 1
    
    return True

def input_guest() :
    global front , rear , queue_size , queue
    
    if front == rear and queue[rear] == None :
        return False
    
    front += 1 
    data = queue[front]
    queue[front] = None
    
    for i in range(front , rear , 1) :
        queue[i] = queue[i + 1]
        queue[i + 1] = None
    front -= 1
    rear -= 1
    
    return data
    
def print_queue() :
    global front , rear , queue_size
    
    for i in range(front + 1 , rear + 1 , 1) :
        print(queue[i] , end = ' , ')
    print()

while True :
    menu = input('1. 대기 손님 입력 // 2. 가게로 손님 들어감 // 3. 대기 중인 손님 출력 \n ')
    
    if menu == 'q' :
        break
    elif menu == '1' :
        guest = input('방문한 손님 수를 입력하시오 : ')
        guest = int(guest)
        if waiting(guest) :
            print('성공적으로 대기 합니다.')
        else :
            print('대기 실패')
    elif menu == '2' :
        data = input_guest()
        if data :
            print('손님이 한명 들어갔습니다.')
        else :
            print('들어갈 손님이 없습니다.')
    elif menu == '3' :
        print_queue()
    else :
        print('입력 오류')
```

```python
# queue : 선입선출 : 입구와 출구가 존재
# a ~ z 까지의 문자중 랜덤으로 선택해서 queue 를 작성
# 전체 큐의 갯수는 임으로 작성
# 다 작성이 끝난 후에 추출 할 숫자 입력해서 해당 숫자의 큐 자료 출력 후
# 자료 이동 후 큐의 자료 상태 확인
import random

def input_queue(data) :
    global front , rear , queue , queue_size
    
    if rear == queue_size - 1 :
        return 
    
    rear += 1
    queue[rear] = data
    
def output_queue(count) :
    global front , rear , queue , queue_size
    
    if count > rear - front :
        print('출력 갯수의 범위를 넘었습니다.')
        return False
    
    data = []
    
    for _ in range(count) :
        front += 1
        data.append(queue[front])
        queue[front] = None
        
        for i in range(0, rear , 1) :
            queue[i] = queue[i + 1]
            queue[i + 1] = None
        rear -= 1
        front = -1
        
    
    return data

def print_queue() :
    global front , rear , queue_size
    
    for i in range(front + 1 , rear + 1 , 1) :
        print(queue[i] , end = ' , ')
    print()

front = -1
rear = -1

queue_size = random.randint(1, 10)
queue = [None for _ in range(queue_size)]

name = 'abcdefghijklmnopqrstuvwxyz'

for i in range(queue_size) :
    data = random.choice(name)
    input_queue(data)

print_queue()
    
count = int(input('출력을 원하는 갯수를 입력하시오 : '))
data = output_queue(count)
if data :
    print('추출한 데이터 값은 : ', *data)

print_queue()
```
