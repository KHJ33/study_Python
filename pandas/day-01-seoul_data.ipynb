```python
# 서울시 cctv 설치 현황을 파일에서 가져와서 cctv_seoul 데이터프레임 생성
# 2번째 줄의 내용을 컬럼으로
import pandas as pd

file_path = './dataset/서울시CCTV설치운영현황(자치구)_년도별_211231기준.csv'
cctv_seoul = pd.read_csv(file_path , encoding = 'cp949' , skiprows = 1)

# 구분을 구명, 구분이 '계'인 자료는 삭제
cctv_seoul.rename(columns = {'구분':'구이름'}, inplace = True)
cctv_seoul.set_index('구이름', inplace = True)
cctv_seoul.drop('계',axis = 0 , inplace = True)
cctv_seoul
```

```python
# cctv 데이터를 파일로 저장
cctv_seoul.to_csv('./dataset/cctv.csv' , encoding = 'utf-8-sig' , index = None)
```
