```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# "시도별 전출입 인구수.xlsx"
df = pd.read_excel('./dataset/시도별 전출입 인구수.xlsx')
df.head(25)
```

```python
# 전출지가 '서울특별시'인 자료만 확인
df[ df['전출지별'] == '서울특별시' ]
```

```python
#  전출지별에 Nan을 해결 : df.fillna(값, method = 'ffill')
df.fillna(method='ffill',inplace=True)  # 자료가 Nan인 경우 앞의 자료로 대체
df[ df['전출지별'] == '서울특별시' ]
```

```python
# 전입지별이 서울특별시 이고, 전입지별이 서울특별시가 아닌 자료 추출
df_seoul = df[ ( df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시' ) ]
df_seoul
```

```python
# 전출지별이 모두 같은 값이므로 전출지별 컬럼 삭제
df_seoul = df_seoul.drop('전출지별', axis=1)
df_seoul
```

```python
# 전입지별을 전입지로 이름 변경
df_seoul.rename(columns = {'전입지별': '전입지'}, inplace=True)
df_seoul
```

```python
# 전입지 컬럼을 인덱스로 지정
df_seoul.set_index('전입지', inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 전입한 인구 분석
df_one = df_seoul.loc['경기도']
df_one
```

```python
plt.plot(df_one.index , df_one.values)
```

```python
import matplotlib as mlt

mlt.rc('font', family='AppleGothic')  # 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False  # -기호 사용 가능
```

```python
# 서울에서 경기도로 이동한 인구의 변화를 라인 그래프로 그리기
plt.figure(figsize=(10,5))
plt.style.use('fivethirtyeight')

plt.plot(df_one, color='green', marker='^')

#   title
plt.title('서울 -> 경기도 이동 인구수')
plt.legend(labels = ['서울 -> 경기'])
plt.xlabel('연 도')
plt.xticks(rotation=75)
plt.ylabel('인구 수')

# y축 범위 지정 (최소값, 최대값)
plt.ylim(50000, 800000)

# 주석 표시 - 화살표
plt.annotate('',
             xy=(20, 620000),       #화살표의 머리 부분(끝점)
             xytext=(2, 290000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='r', lw=3), #화살표 서식
             )

plt.annotate('',
             xy=(47, 450000),       #화살표의 머리 부분(끝점)
             xytext=(30, 580000),   #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  #화살표 서식
             )

# 주석 표시 - 텍스트
plt.annotate('인구이동 증가(1970-1995)',  #텍스트 입력
             xy=(10, 380000),            #텍스트 위치 기준점
             rotation=30,                #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=10,                #텍스트 크기
             )

plt.annotate('인구이동 감소(1995-2017)',  #텍스트 입력
             xy=(40, 500000),            #텍스트 위치 기준점
             rotation=-12,               #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=10,                #텍스트 크기
             )

plt.show()
```

```python
# 경기도 인구 이동의 그래프를 하나의 페이지에 2개를 1열에 1개
# fig.add_subplot(행의 갯수, 열의 갯수, 순번)
fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(df_one, color = 'r', label = '서울->경기')
ax1.legend()
ax1.set_title('서울->경기')

ax2.plot(df_one, color = 'b')
ax2.set_title('서울->경기')

ax1.set_xticklabels(df_one.index , rotation = 75)
# x, y 축 눈금 라벨 크기
ax1.tick_params(axis = 'x' , labelsize = 8)
ax1.tick_params(axis = 'y' , labelsize = 15)
ax2.set_xticklabels(df_one.index, rotation = 90)
ax1.tick_params(axis = 'x' , labelsize = 10)

plt.show() # 변경사항 저장후 보여주는것
```

```python
# 서울시에서 충남, 경북, 강원도로 이동한 인구수의 그래프를 그리고자 함
df = pd.read_excel('./dataset/시도별 전출입 인구수.xlsx')
df.fillna(method = 'ffill', inplace = True)
df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul = df_seoul.drop(columns = ['전출지별'])
df_seoul = df_seoul.set_index('전입지별')
```

```python
df_three = df_seoul.loc[['충청남도','경상북도','강원도']]
df_three
```

```python
# 그래프 그리기
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(1,1,1)

ax.plot(df_three.columns, df_three.loc['충청남도',:], color = 'blue', label = '서울 -> 충남' , marker = 'o')
ax.plot(df_three.columns, df_three.loc['경상북도',:], color = 'orange', label = '서울 -> 경북', marker = '.')
ax.plot(df_three.columns, df_three.loc['강원도',:], color = 'green', label = '서울 -> 강원도', marker = '^')

ax.legend()
ax.set_title('서울에서 충남, 경북, 강원 이동 인구', size = 15)
ax.set_xlabel('기간')
ax.set_ylabel('이동 인구 수')

ax.set_xticklabels(df_three.columns , rotation = 75, size = 10)

plt.show()
```

```python
# 전라북도 자료를 추가해서 df_four를 생성
df_four = df_three.copy()
df_four.loc['전라북도'] = df_seoul.loc['전라북도']
df_four
```

```python
fig = plt.figure(figsize = (15,15))
ax1 = fig.add_subplot(2,2,1) # 충남
ax2 = fig.add_subplot(2,2,2) # 경북
ax3 = fig.add_subplot(2,2,3) # 강원
ax4 = fig.add_subplot(2,2,4) # 전북

col_years = df_four.columns
ax1.plot(col_years, df_four.loc['충청남도'], color = 'r', label = '서울->충남')
ax2.plot(col_years, df_four.loc['경상북도'], color = 'g', label = '서울->경북')
ax3.plot(col_years, df_four.loc['강원도'], color = 'skyblue', label = '서울->강원')
ax4.plot(col_years, df_four.loc['전라북도'], color = 'b', label = '서울->전북')

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()

ax1.set_title('서울 -> 충남 이동인구 수')
ax2.set_title('서울 -> 경북 이동인구 수')
ax3.set_title('서울 -> 강원 이동인구 수')
ax4.set_title('서울 -> 전북 이동인구 수')

plt.show()
```

```python
# pandas plot() 함수로 면적 그래프 작성
# x축에 눈금 -> 정수형으로 => df_four -> index가 도시, 연도가 컬럼
# df_four 를 전치행렬로 바꿈 : df_four = df_four.T
df_four = df_four.T
df_four
```

```python
df_four.index = df_four.index.map(int)
df_four.index
```

```python
# 면적 그래프는 -> 인덱스가 x 축으로 컬럼별로 면적 그래프를 그림
df_four.plot(kind = 'area' , stacked = False , alpha = 0.3 , figsize = (20,10))

plt.title('서울 -> 충남, 경북, 강원, 전북의 이동 인구 수')
plt.xlabel('기간')
plt.ylabel('이동 인구 수')

plt.show()
```

```python
# 면적 그래프는 -> 인덱스가 x 축으로 컬럼별로 면적 그래프를 그림
ax = df_four.plot(kind = 'area' , stacked = False , alpha = 0.3 , figsize = (20,10))

ax.set_title('서울 -> 충남, 경북, 강원, 전북의 이동 인구 수')
ax.set_xlabel('기간')
ax.set_ylabel('이동 인구 수')
ax.legend()
plt.show()
```
