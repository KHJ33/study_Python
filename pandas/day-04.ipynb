```python
# titanic의 데이터를 가져와서 df에 저장한 후
# Nan 이 있는 컬럼을 전처리 진행
# Nan 데이터의 갯수가 300 이상인 컬럼은 삭제 -> dropna
# age 컬럼의 경우 가장 작은 나이와 가장 많은 나이의 평균 나이로 대체 -> fillna(값)
# embark_town 은 이전에 나오는 값으로 대체 -> fillna(method = )
# embarked 는 가장 빈번하게 나오는 컬럼의 값으로 대체 -> fillna(값)

import pandas as pd
import seaborn as sns
import math

df = sns.load_dataset('titanic')
# df.info()

print(891 - df[['age','embarked','embark_town','deck']].count())

df = df.dropna(axis = 1 , thresh = 300)
# df.info()

df['age'].fillna(math.trunc(df['age'].max() + df['age'].min()) / 2 , inplace = True)

df['embark_town'].fillna(method = 'ffill', inplace = True)

df['embarked'].fillna(df['embarked'].value_counts().idxmax(), inplace = True)

df.info()
```

## 범주형 자료 처리

```python
# auto-mpg.csv 파일을 읽어 df 에 저장한 후
# horesepower 를 구간 분할해서 구간 변수로 지정
# horespower 데이터 타입이 숫자이어야 함

df = pd.read_csv('./dataset/auto-mpg.csv', header = None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

df.horsepower.replace('?', np.nan , inplace = True)
df.dropna(subset = ['horsepower'], axis = 0 , inplace = True)
df['horsepower'] = df['horsepower'].astype('float')
```

```python
# np.histogram() 함수를 사용하여 구간의 경계 값 구한
cnt, bin_dividers = np.histogram(df['horsepower'], bins = 3)
print(cnt, bin_dividers)
```

```python
# pandas 의 cut() 함수를 이용하여 구간 분리
bin_names = ['저출력','보통출력','고출력']

df['hp_bin'] = pd.cut(x = df['horsepower'],
                      bins = bin_dividers,
                      labels = bin_names , 
                      include_lowest = True)

df[['horsepower','hp_bin']].head(10)
```

## 더미 변수 

```python
# 더미 변수 생성 -> get_dummies(컬럼)
pd.get_dummies(df['hp_bin']) # 원핫 인코딩

# pd.get_dummies(df['hp_bin'], prefix = 'hp_bin')
```

## 더미 함수는 category 타입을 변경해준다

```python
df_1 = df[['horsepower','origin','hp_bin']].copy()
df_1

# origin 을 1 : USA, 2 : EU , 3 : JPN 으로 변경하고 origin 의 데이터 타입을 category 로 변경
df_1.replace({1:'USA',2:'EU',3:'JPN'}, inplace = True)
df_1['origin'] = df_1['origin'].astype('category')
df_1.info()
```

```python
df2 = pd.get_dummies(df_1, drop_first = True) # 첫번째 하나가 없어도 예측이 가능하니 첫번째를 뺴준다
df2
```

```python
# titanic 의 age 컬럼을 유아, 청소년, 중년, 장년으로 나누어서 분류하는 df['age_gubun']
# dummy 로 출력
# age 의 Nan 값은 평균 값으로 대체(소수점 미만 버림)
# 구간을 나누고, category 로 변경 후 dummy 함수 적용

import pandas as pd
import seaborn as sns
import math
import numpy as np

titanic = sns.load_dataset('titanic')
# sex -> category 로 변경 , alive -> category 로 변경
# 숫자와 category 만 추출하여 데이터 프레임으로 생성 후 dummy 함수

titanic['age'] = titanic['age'].fillna(math.trunc(titanic['age'].mean()))

cnt, bin_dividers = np.histogram(titanic['age'], bins = 4)
print(cnt , bin_dividers)
bin_names = ['유아','청소년','중년','장년']

titanic['age_bin'] = pd.cut(x = titanic['age'],
                      bins = bin_dividers,
                      labels = bin_names , 
                      include_lowest = True)

# titanic['age_bin']
titanic['sex'] = titanic['sex'].astype('category')
titanic['alive'] = titanic['alive'].astype('category')

titanic.info()

df = pd.DataFrame()

for col in titanic.columns:
    if (titanic[col].dtype == 'int64') or (titanic[col].dtype == 'float64') or (titanic[col].dtype.name == 'category') :
        df[col] = titanic[col].copy()

# df = titanic[['sex','alive','age_bin']].copy()
df
```

# 시계열 데이터

```python
# 시계열 데이터
# stock-data.csv 파일 -> header 존재
# date -> pd.to_datetime(바꿀자료) => 'new_data' -> datetime 형식으로 변경
# 기존의 date 컬럼 삭제, new_date 를 인덱스로

df = pd.read_csv('./dataset/stock-data.csv')
df['new_Date'] = pd.to_datetime(df['Date'])
df.info()

# 기존 컬럼 삭제
df.drop('Date', axis = 1 , inplace = True)
df.set_index('new_Date', inplace = True)
df
```

```python
df.reset_index(inplace = True)
```

```python
df['Year'] = df['new_Date'].dt.year
df['Month'] = df['new_Date'].dt.month
df['Day'] = df['new_Date'].dt.day
df.head()
```

```python
# 기간 객체로 변환 => to_period
df['date_yr'] = df['new_Date'].dt.to_period(freq = 'A')
df['date_ym'] = df['new_Date'].dt.to_period(freq = 'M')
df.head()
```

```python
df.set_index('new_Date', inplace = True)
df
```
