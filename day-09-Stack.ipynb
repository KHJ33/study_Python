```python
# stack : FILO , push , pop , top
# 처음 넣은 자료가 제일 마지막에 나옴

stack = [None , None, None, None , None]
top = -1

# 커피를 추가
top += 1
stack[top] = '커피'

# 녹차 추가
top += 1
stack[top] = '녹차'

# 꿀물 추가
top += 1
stack[top] = '꿀물'

#stack 의 자료 출력
for i in range(len(stack) - 1 , -1 , -1) :
    print(stack[i])
```

```python
# 스택 초기화
# push() -> stack 이 full 인지 확인 , 자료 삽입
# pop() -> stack 의 자료가 있는지 확인 , 자료 추출
# 스택의 활용 => 웹페이지 이전 페이지로 돌아갈때

def push(data) :
    global top , stack_size , stack
    
    # stack 이 Full 인지 확인
    if isStackFull() :
        print('스택이 꽉 찼습니다.')
        return
    top += 1
    stack[top] = data
    
def isStackFull() :
    global top , stack_size , stack
    
    if top == stack_size - 1 : # stack 이 Full 로 찼으면 True 반환
        return True
    else :
        return False
    
def pop() :
    global top , stack_size , stack
    
    # stack 이 Full 인지 확인
    if isStackEmpty() :
        print('스택이 비어 있습니다.')
        return None
    
    data = stack[top]
    stack[top] = None
    top -= 1
    return data
    
def isStackEmpty() :
    global top
    
    if top == -1 : # stack 이 비어 있으면 True 반환
        return True
    else :
        False

def print_stack() :
    global stack
    
    for i in range(len(stack) - 1 , -1 , -1) :
        print(stack[i])
        
def peek() :
    global top, stack
    
    if top == -1 :
        print('자료가 없습니다.')
        return False
    else :
        return stack[top]

# stack 의 사이즈 지정
stack_size = int(input('스택 사이즈를 입력하세요 : '))
stack = [ None for _ in range(stack_size)]
top = -1
        
if __name__ == '__main__' :
    while True :
        menu = input('1 : 자료 삽입 , 2 : 자료 추출 , 3 : 스택의 자료 , 4 : 스택의 마지막 자료 확인 , 9 : 종료\n')

        if menu == '9' :
            break

        if menu == '1' :
            data = input('삽입할 자료 입력 : ')
            push(data)

        elif menu == '2' :
            data = pop()
            if data :
                print('추출한 자료는 ', data)

        elif menu == '3' :
            print('스택의 자료 출력 :')
            print_stack()

        elif menu == '4' :
            data = peek()
            if data :
                print('마지막 자료는 ', data)

        else :
            print('메뉴 선택 오류 입니다.')
```
